## 시간: 3시간

### ▶ 1시간차: 정적 크롤링 기초 기술 자체 이해
- 정적 크롤링 vs 동적 크롤링 구분
- 가장 기본 파이썬 requests + BeautifulSoup4 설치
- 기존 플랫폼 (국내): [https://www.anuga.com/]
- 크롤링 예제:
```python
import requests
from bs4 import BeautifulSoup

url = "https://www.anuga.com/"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
print(soup.title.text)
```
- `<div>`, `<a>`, `<img>` 속성 검색
- `.select()`, `.get_text()`, `['href']`, `['src']` 사용
